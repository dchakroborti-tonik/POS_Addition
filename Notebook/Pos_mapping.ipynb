```python
# Mapping function to apply conditions
def map_brand_category(x):
    if 'Homeworks' in x:
        return 'Homeworks'
    elif 'Workstation' in x:
        return 'Workstation'
    elif 'Jhunpyo' in x:
        return 'Jhunpyo'
    elif 'Xtreme' in x:
        return 'Xtreme'
    elif 'Affordables' in x:
        return 'Affordables'
    elif 'Faith Glorious' in x:
        return 'Faith Glorious'
    elif 'FC Home' in x:
        return 'FC Home'
    elif 'A2RJ' in x:
        return 'A2RJ'
    elif 'ESLA' in x:
        return 'ESLA'
    elif 'JTG' in x:
        return 'JTG'
    elif 'KUM Home' in x:
        return 'KUM Home'
    elif 'Willy and Sons' in x:
        return 'Willy & Sons'
    elif 'PHPS' in x:
        return 'PHPS'
    elif 'Junliz' in x:
        return 'Junliz'
    elif 'Home Factory' in x:
        return 'Home Factory'
    elif 'Sogo Home' in x:
        return 'Sogo Home'
    elif 'Home Along' in x:
        return 'Home Along'
    elif 'Power Mac Center' in x:
        return 'Power Mac Center'
    elif 'Affordables' in x:
        return 'Affordables '
    else:
        return 'Other'

# Apply the mapping function to create a new 'BrandStores' column
df['BrandStores'] = df['POS NAME'].apply(map_brand_category)

# Display the updated DataFrame
df
```
